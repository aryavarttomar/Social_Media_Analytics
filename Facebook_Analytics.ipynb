###Initial Inspection of Facebook_Post_Engagement###
Viewing the first 5 rows revealed a duplicate header row at the top, which should be removed. Additionally, the RowHash column has a missing value in its first cell, indicating it is misaligned. To fix this, we will first shift the cells in RowHash column down by one row to correctly align it with the rest of the data, and then remove the duplicate header row.

```python
# Shift the 'RowHash' column down by one row to align it properly with the rest of the data
Facebook_Post_Engagement['RowHash'] = Facebook_Post_Engagement['RowHash'].shift(1)

# Display the first few rows to confirm the shift
Facebook_Post_Engagement.head(5)
```

```python
# Drop the first row (index 0), which is a duplicate header row
Facebook_Post_Engagement = Facebook_Post_Engagement.drop(index=0).reset_index(drop=True)

# Display the first few rows to confirm the removal
Facebook_Post_Engagement.head(5)
```

```python
# Step 2: View the datatypes of each column in Facebook_Profile_Overview
print(Facebook_Post_Engagement.dtypes)
# Step 3: Check for any null (NaN) values in Facebook_Profile_Overview
print(Facebook_Post_Engagement.isnull().sum())
```

 ### Convert Column Data Types in Facebook_Post_Engagement ###

We converted the column Post creation date to datetime64[ns] using the day-first format (DD/MM/YYYY) to support accurate time-based analysis, such as plotting engagement trends over time.

Next, we identified several columns that represent numerical engagement metrics (e.g., impressions, reach, reactions, shares) but were incorrectly read as object types. These columns were converted to numeric (Int64) to enable proper aggregation, filtering, and visualization.

Using errors='coerce' ensured that any non-numeric values or formatting inconsistencies were safely handled by converting them to NaN, rather than breaking the code.

```python
# Convert 'Post creation date' to datetime with day-first format
Facebook_Post_Engagement['Post creation date'] = pd.to_datetime(
    Facebook_Post_Engagement['Post creation date'], dayfirst=True, errors='coerce'
)


# Columns to convert to int
int_columns = [
    'Post impressions', 'Post organic reach', '% of reach from paid', '% of reach from organic',
    'Post reach', 'Shares on posts', 'Total post reactions', 'Post photo views', 'Video views',
    'Comments on posts', 'Shares on shares', 'Organic video views', 'Post link clicks'
]

# Convert specified columns to numeric (int)
for col in int_columns:
    Facebook_Post_Engagement[col] = pd.to_numeric(Facebook_Post_Engagement[col], errors='coerce').astype('Int64')

# Show updated datatypes
Facebook_Post_Engagement.dtypes
```

```python
# Step 3: Check for null values in Facebook_Post_Engagement after cleaning
Facebook_Post_Engagement.isnull().sum()
```

```python
# Step 4: Count zero values in each column of Facebook_Post_Engagement
zero_value_counts_post = (Facebook_Post_Engagement == 0).sum()

zero_value_counts_post
```

 ### Zero Values Analysis – Facebook_Post_Engagement###
Several engagement-related columns contain a large number of zero values, particularly:

% of reach from paid

Shares on posts, Shares on shares

Video views, Organic video views

Post link clicks, Comments on posts, etc.

These zeros indicate posts that received no engagement on specific actions, which is expected in social media datasets and should not be removed. Instead, they are valuable for comparing high vs. low performing posts.

###Conclusion:###
Zero values here are meaningful and will help analyze which posts performed well versus those that didn't.
