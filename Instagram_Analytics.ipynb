### **2. Basic Analysis** ###



### Engagement Rate Calculation ###
#### Total Engagements: ####
We compute total engagements using the equation:

#### Total Engagements = ∑(Likes+Comments+Shares) ####

#### Total Impressions: ####
We sum all the Media reach to get the total number of times posts were seen:
#### Total Impressions = ∑(Media Impressions) ####

#### Engagement Rate: ####
We calculate the engagement rate by dividing total engagements by total impressions:

#### Engagement Rate = Total Engagements/Total Impressions ####

#### Convert to Percentage: ####
We multiply the engagement rate by 100 and round to two decimal places to express it as a percentage.

This provides a single, clear KPI showing how often users interacted with content relative to how many times it was shown.

```python
Instagram_Post_Engagement["Total Engagement"] = (
    Instagram_Post_Engagement["Like count"] +
    Instagram_Post_Engagement["Comments count"] +
    Instagram_Post_Engagement["Shares"]
)


# Calculate overall total engagements and total impressions
total_engagements = Instagram_Post_Engagement["Total Engagement"].sum()
total_impressions = Instagram_Post_Engagement["Media reach"].sum()

# Calculate engagement rate
engagement_rate = total_engagements / total_impressions

# Convert engagement rate to percentage
engagement_rate_percent = round(engagement_rate * 100, 2)
print(engagement_rate_percent)
```

The overall engagement rate for Instagram posts is approximately: 4.68%
This means that only about 4.68% of all users who saw a post interacted with it (liked, commented, or shared)

```python
# Calculate engagement rate per post and save to new column
Instagram_Post_Engagement["Engagement_Rate_Per_Post"] = (
    Instagram_Post_Engagement["Total Engagement"] /
    Instagram_Post_Engagement["Media reach"]
).round(4)

# Display relevant columns to confirm
Instagram_Post_Engagement[[
    "Date", "Total Engagement", "Media reach", "Engagement_Rate_Per_Post"
]].head()
```

### Predicting Post Performance Using a Threshold-Based Model ###
In this step, we created a simple rule-based model to label each post as either “High” or “Low” performing, based on total engagement (likes + comments + shares).

Explanation:

### Defined a Function: ###
We wrote a function that classifies a post as "High" if its total engagement is greater than or equal to a threshold, and "Low" otherwise.

### Used Median as the Threshold: ###
The threshold was set to the median of total engagement values across all posts.

### Applied the Function: ###
We used .apply() to run this logic row by row and created a new column called "Performance Prediction" in the dataset.

### Q Why We Chose Median Instead of Mean: ###

### Median is more robust to outliers: ###
A few viral posts with extremely high engagement could skew the mean, making it an unfair threshold.
### Median gives a balanced split: ###
About half the posts will be labeled "High" and half "Low," giving a clearer view of typical post performance.
This method provides a quick and interpretable way to segment content based on past engagement,  ideal for reporting or guiding future content strategy.

```python
# Define a simple threshold-based prediction function
def predict_performance(row, threshold=Instagram_Post_Engagement["Total Engagement"].median()):
    """
    Predicts if a post is 'High' or 'Low' performing based on total engagement.
    Uses the median engagement as a threshold unless specified otherwise.
    """
    return "High" if row["Total Engagement"] >= threshold else "Low"

# Apply the function to the DataFrame
Instagram_Post_Engagement["Performance Prediction"] = Instagram_Post_Engagement.apply(predict_performance, axis=1)

# Display first few rows with the new prediction column
Instagram_Post_Engagement[["Date", "Media product type", "Total Engagement", "Performance Prediction"]].head()
```
